```python
import json

filename = 'SLAM_MITb'

# Extract associated simulation data
with open(f"Results/{filename}.json") as f:
    data_simulations = json.load(f)
```

```python
limits = []
random_false = []
random_true = []
random_exclusion_false = []
random_exclusion_true = []
fixed_false = []
fixed_true = []
residual_false = []
residual_true = []
for limit, data in data_simulations.items():
    limits.append(int(limit))

    random_false.append(data['stream_results']['random_False'])
    random_true.append(data['stream_results']['random_True'])
    random_exclusion_false.append(data['stream_results']['random-exclusion_False'])
    random_exclusion_true.append(data['stream_results']['random-exclusion_True'])
    fixed_false.append(data['stream_results']['fixed_False'])
    fixed_true.append(data['stream_results']['fixed_True'])
    residual_false.append(data['stream_results']['residual_False'])
    residual_true.append(data['stream_results']['residual_True'])
```

```python
print(random_false)
print(random_true)
print(random_exclusion_false)
print(random_exclusion_true)
print(fixed_false)
print(fixed_true)
print(residual_false)
print(residual_true)
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))

# Plot the scatter plot

# Plot the Pareto frontier line
plt.plot(limits, random_false, color='red', label='Random')
plt.plot(limits, random_true, color='red', linestyle='--',label='Random - Cache')
plt.plot(limits, random_exclusion_false, color='blue', label='Random (Exclusion)')
plt.plot(limits, random_exclusion_true, color='blue', linestyle='--', label='Random (Exclusion) - Cache')
plt.plot(limits, fixed_false, color='green', label='Fixed')
plt.plot(limits, fixed_true, color='green', linestyle='--', label='Fixed - Cache')
plt.plot(limits, residual_false, color='orange', label='Residual')
plt.plot(limits, residual_true, color='orange', linestyle='--', label='Residual - Cache')

plt.legend()

# Add labels and title
plt.xlabel('Maximum Resource Utilization (as a Percentage of Total Available)', fontsize=10)  # Adjusted font size here
plt.ylabel('Number of Data Streams Required until Convergence', fontsize=10)  # Adjusted font size here

plt.title('MITb Dataset (N = 808, k = 832)', fontsize=12)  # Adjusted font size here

# Show plot
plt.grid(True)

plt.savefig(f"MITb-dataset.png") 

plt.show()
```
