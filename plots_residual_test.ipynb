### Extract Data

```python
import json

filename = 'SLAM_MITb_residual_test_mae'

# Extract associated simulation data
with open(f"Results/{filename}.json") as f:
    data_simulations = json.load(f)
```

```python
limits = []

random = []
fixed = []
residual = []

residual_TMAX_PREV = []
residual_TMAX_CURR = []
residual_TnP_PREV = []
residual_TnP_CURR = []

residual_TMAX_PREV_scaling = []
residual_TMAX_CURR_scaling = []
residual_TnP_PREV_scaling = []
residual_TnP_CURR_scaling = []

for limit, data in data_simulations.items():
    limits.append(int(limit))
    fixed.append(data['stream_results']['fixed_True'])
    residual.append(data['stream_results']['residual_True'])
    residual_TMAX_PREV.append(data['stream_results']['residual-TMAX-PREV_True'])
    residual_TMAX_CURR.append(data['stream_results']['residual-TMAX-CURR_True'])
    residual_TnP_PREV.append(data['stream_results']['residual-TnP-PREV_True'])
    residual_TnP_CURR.append(data['stream_results']['residual-TnP-CURR_True'])


    residual_TMAX_PREV_scaling.append(data['stream_results']['residual-TMAX-PREV-scaling_True'])
    residual_TMAX_CURR_scaling.append(data['stream_results']['residual-TMAX-CURR-scaling_True'])
    residual_TnP_PREV_scaling.append(data['stream_results']['residual-TnP-PREV-scaling_True'])
    residual_TnP_CURR_scaling.append(data['stream_results']['residual-TnP-CURR-scaling_True'])
```

```python
padding_residual_TMAX_PREV = []
padding_residual_TMAX_CURR = []
padding_residual_TnP_PREV = []
padding_residual_TnP_CURR = []

padding_residual_TMAX_PREV_scaling = []
padding_residual_TMAX_CURR_scaling = []
padding_residual_TnP_PREV_scaling = []
padding_residual_TnP_CURR_scaling = []

for limit, data in data_simulations.items():
    padding_residual_TMAX_PREV.append(data['padding_results']['residual-TMAX-PREV_True'])
    padding_residual_TMAX_CURR.append(data['padding_results']['residual-TMAX-CURR_True'])
    padding_residual_TnP_PREV.append(data['padding_results']['residual-TnP-PREV_True'])
    padding_residual_TnP_CURR.append(data['padding_results']['residual-TnP-CURR_True'])


    padding_residual_TMAX_PREV_scaling.append(data['padding_results']['residual-TMAX-PREV-scaling_True'])
    padding_residual_TMAX_CURR_scaling.append(data['padding_results']['residual-TMAX-CURR-scaling_True'])
    padding_residual_TnP_PREV_scaling.append(data['padding_results']['residual-TnP-PREV-scaling_True'])
    padding_residual_TnP_CURR_scaling.append(data['padding_results']['residual-TnP-CURR-scaling_True'])
```

```python
print(f"padding_residual_TMAX_PREV = {padding_residual_TMAX_PREV}")
print(f"padding_residual_TMAX_CURR = {padding_residual_TMAX_CURR}")
print(f"padding_residual_TnP_PREV = {padding_residual_TnP_PREV}")
print(f"padding_residual_TnP_CURR = {padding_residual_TnP_CURR}")

print(f"padding_residual_TMAX_PREV_scaling = {padding_residual_TMAX_PREV_scaling}")
print(f"padding_residual_TMAX_CURR_scaling = {padding_residual_TMAX_CURR_scaling}")
print(f"padding_residual_TnP_PREV_scaling = {padding_residual_TnP_PREV_scaling}")
print(f"padding_residual_TnP_CURR_scaling = {padding_residual_TnP_CURR_scaling}")
```

### Compare New Methods
