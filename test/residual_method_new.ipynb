### Linear Regression - prev T max

```python
async_convergence_threshold = 1 * 10**-5

sum_of_iterations = 0
num_iterations = ASYNC_ITER

# Track the valid iterations for averaging
valid_iterations = 0

for it in range(num_iterations):
    while True:
        # Run the hardware-accelerated residual function
        P_i, mu_i, iteration, hardware_t_max_lst, PADDING_ANALYTICS = run_GaBP_HARDWARE_ACCELERATED_RESIDUAL_T_nP(A, b, test_method=False, SCALING=False, SHUFFLE_STREAM=False, CURR_T_MAX=False, caching=True, node_updates_per_pe=NODE_UPDT_PE, number_pes=PEs, TRUE_MEAN=final_mean, max_iter=10000, mae=True, convergence_threshold=async_convergence_threshold, show=True)
        
        # Check if iteration is valid
        if iteration <= 10000:
            sum_of_iterations += iteration
            valid_iterations += 1
            break  # Exit the while loop and proceed to the next for loop iteration
        else:
            pass
            # print(f"Iteration {it+1} exceeded 10000 steps. Repeating...")

    print(f"-------------- ITERATION = {it+1} => Streams = {iteration} --------------")

# Calculate the average of valid iterations
if valid_iterations > 0:
    average_iterations = sum_of_iterations / valid_iterations
else:
    average_iterations = 0

print(f"AVE. ASYNC ITERATIONS = {average_iterations}")

print(PADDING_ANALYTICS)
print(sum(PADDING_ANALYTICS))
print(sum(PADDING_ANALYTICS)/len(PADDING_ANALYTICS))
```

```python
# Generating x values
x_val_prev_t_max_lst = np.array([i + 1 for i in range(len(prev_t_max_lst))])
x_val_curr_t_max_lst = np.array([i + 1 for i in range(len(curr_t_max_lst))])

# Performing linear regression in log space (line of best fit)
coef_prev = np.polyfit(x_val_prev_t_max_lst[2:], np.log(prev_t_max_lst[2:]), 1)
coef_curr = np.polyfit(x_val_curr_t_max_lst[2:], np.log(curr_t_max_lst[2:]), 1)

# Exponentiating to get back to linear scale for the coefficients
a_prev = np.exp(coef_prev[1])   # intercept
b_prev = coef_prev[0]           # slope

a_curr = np.exp(coef_curr[1])   # intercept
b_curr = coef_curr[0]           # slope

# Creating the plot with a logarithmic y-axis
plt.figure(figsize=(10, 6))

# Plotting data points for prev_t_max_lst[2:]
plt.semilogy(x_val_prev_t_max_lst[2:], prev_t_max_lst[2:], marker='o', linestyle='-', color='b', label='prev_t_max_lst[2:]')

# Plotting line of best fit for prev_t_max_lst[2:]
plt.semilogy(x_val_prev_t_max_lst[2:], np.exp(coef_prev[1]) * np.exp(coef_prev[0] * x_val_prev_t_max_lst[2:]), linestyle='--', color='b', label='Line of Best Fit (prev_t_max_lst[2:])')

# Plotting data points for curr_t_max_lst[2:]
plt.semilogy(x_val_curr_t_max_lst[2:], curr_t_max_lst[2:], marker='o', linestyle='-', color='r', label='curr_t_max_lst[2:]')

# Plotting line of best fit for curr_t_max_lst[2:]
plt.semilogy(x_val_curr_t_max_lst[2:], np.exp(coef_curr[1]) * np.exp(coef_curr[0] * x_val_curr_t_max_lst[2:]), linestyle='--', color='r', label='Line of Best Fit (curr_t_max_lst[2:])')

# Adding title and labels
plt.title('Plot of prev_t_max_lst[2:] and curr_t_max_lst[2:] with Logarithmic Y-axis and Lines of Best Fit')
plt.xlabel('Iterations')
plt.ylabel('T_max (log)')

# Adding legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()

# Printing the equations of the lines of best fit
print(f"Equation of the line of best fit for prev_t_max_lst[2:]: T_max = {a_prev:.2f} * e^({b_prev:.2f} * Iterations)")
print(f"Equation of the line of best fit for curr_t_max_lst[2:]: T_max = {a_curr:.2f} * e^({b_curr:.2f} * Iterations)")
```
